```python
import os

def create_basketball_scraper_structure():
    """יוצר את מבנה התיקיות והקבצים של basketball_scraper"""
    
    # מבנה התיקיות והקבצים
    structure = {
        'basketball_scraper': {
            'files': ['config.py', 'main.py', 'README.md'],
            'models': {
                'files': ['__init__.py', 'data_models.py']
            },
            'scrapers': {
                'files': ['__init__.py', 'ibasketball.py']
            },
            'utils': {
                'files': ['__init__.py', 'helpers.py']
            },
            'data': {
                'files': ['leagues.csv', 'teams.csv', 'players.csv'],
                'leumit': {
                    'files': ['team_names.csv', 'games_schedule.csv'],
                    'leumit_games': {
                        'files': []
                    }
                },
                'artzit': {
                    'files': []
                }
            },
            'logs': {
                'files': ['update_log.txt']
            }
        }
    }
    
    def create_structure(base_path, structure_dict):
        """פונקציה רקורסיבית ליצירת תיקיות וקבצים"""
        for key, value in structure_dict.items():
            if key == 'files':
                # יצירת קבצים
                for filename in value:
                    filepath = os.path.join(base_path, filename)
                    with open(filepath, 'w', encoding='utf-8') as f:
                        if filename == 'README.md':
                            f.write('# Basketball Scraper\n\nמערכת לגזירת נתוני כדורסל\n')
                        elif filename.endswith('.py'):
                            f.write('# -*- coding: utf-8 -*-\n')
                        elif filename.endswith('.txt'):
                            f.write('')
                    print(f'נוצר קובץ: {filepath}')
            else:
                # יצירת תיקייה
                dir_path = os.path.join(base_path, key)
                os.makedirs(dir_path, exist_ok=True)
                print(f'נוצרה תיקייה: {dir_path}')
                
                # המשך רקורסיבי
                if isinstance(value, dict):
                    create_structure(dir_path, value)
    
    # יצירת המבנה מהתיקייה הנוכחית
    base_dir = os.getcwd()
    create_structure(base_dir, structure)
    
    print('\n✅ מבנה התיקיות נוצר בהצלחה!')
    print(f'📁 נוצר בתיקייה: {os.path.join(base_dir, "basketball_scraper")}')

if __name__ == '__main__':
    create_basketball_scraper_structure()
```
